## 1
Background
We all know about "balancing parentheses" (plus brackets, braces and chevrons) and even balancing characters that are identical.


***Kata***  
***Your challenge in this kata is to write a piece of code to validate that a supplied string is balanced.***

You must determine if all that is open is then closed, and nothing is closed which is not already open!

You will be given a string to validate, and a second string, where each pair of characters defines an opening and closing sequence that needs balancing.

You may assume that the second string always has an even number of characters.

*(Это нужно было решать с помощью стака)*

```python
from collections import deque


def is_balanced(source, caps):
    '''
    Проверить является ли строка сбалансированной,
    удалость решить O(n) 
    '''
    open_to_close = {}
    close_to_open = {}
    deq = deque()
    unique_deq = deque()
    for i in range(0, len(caps), 2):
        open_to_close[caps[i]] = caps[i + 1]
        close_to_open[caps[i + 1]] = caps[i]
    for char in source:
        if char in open_to_close:
            if open_to_close[char] == char:
                try:
                    last_value = unique_deq.pop()
                except IndexError:
                    last_value = ''
                    if last_value != char:
                        if last_value:
                            unique_deq.append(last_value)
                        unique_deq.append(char)
            else:
                deq.append(char)
            continue
            
        elif char in close_to_open:
            try:
                last_value = deq.pop()
            except IndexError:
                return False
            if close_to_open[char] != last_value:
                return False

    return len(deq) == 0 and len(unique_deq) == 0


assert is_balanced('-some value', '--') == False
assert is_balanced('-some value-', '--') == True
assert is_balanced('-+some value+-', '--++') == True
assert is_balanced('((Hello))', '()') == True
```

## 2
Есть список [1, 2, 3, 4, 8, 8, 8, 10] нужно получить индексы первого и последнего числа 8 в этом списке

```python

def get_left_and_right_indexes_of_value_in_arr(arr, value):
    '''
    Есть список [1, 2, 3, 4, 8, 8, 8, 10] нужно получить индексы первого и последнего числа 8 в этом списке
    1 Вариант решения:
        Воспользоваться обычным методом бинарного поиска, получить любое индекс числа, а дальше отсчитывать влево
        и вправо пока это левое и правое число равно нашему
        Почему это не подходит?
        Предположим у нас такой случай [1, 2, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 10], 
        в таком случай у нас сложность не O(log(n)), а O(n)
    2 Вариант решения:
        Дважды воспользоваться бинарным поиском, но подкоректировать условие, смотреть на левое/правое соседнее число
        в таком случай у нас всегда сложность O(2log(n))
    '''
    start = 0
    end = len(arr) - 1

    while start <= end:
        mid_index = (start + end) // 2
        mid_value = arr[mid_index]
        if value < mid_value:
            end = mid_index - 1
        # если в "массивe" мы дошли до крайнего, то этот индекс нам и нужен
        if mid_index == end:
            right_index = end
            break
        if value > mid_value or (value == mid_value and value == arr[mid_index + 1]):
            start = mid_index + 1
            continue
        if value == mid_value:
            right_index = mid_index    
            break

    start = 0
    end = len(arr) - 1

    while start <= end:
        mid_index = (start + end) // 2
        mid_value = arr[mid_index]
        if value > mid_value:
            start = mid_index + 1
        # если в "массивe" мы дошли до крайнего, то этот индекс нам и нужен
        if mid_index == start:
            left_index = mid_index
            break
        if value < mid_value or (value == mid_value and value == arr[mid_index - 1]):
            end = mid_index - 1
            continue
        if value == mid_value:
            left_index = mid_index    
            break

    return [left_index, right_index]


arr = [i for i in range(10 ** 3)]
arr.insert(100, 100)
arr.insert(101, 100)

# на большом примере
assert get_left_and_right_indexes_of_value_in_arr(arr, 100) == [100, 102]
# одно число слева
assert get_left_and_right_indexes_of_value_in_arr([8, 10, 12], 8) == [0, 0]
# одно число посередине
assert get_left_and_right_indexes_of_value_in_arr([1, 2, 8, 10], 8) == [2, 2]
# несколько чисел посередине
assert get_left_and_right_indexes_of_value_in_arr([1, 8, 8, 8, 8, 10], 8) == [1, 4]
# несколько чисел справа
assert get_left_and_right_indexes_of_value_in_arr([1, 6, 7, 8, 8, 8, 8, 8], 8) == [3, 7]
# несколько чисел слева
assert get_left_and_right_indexes_of_value_in_arr([8, 8, 8, 8, 12, 15, 18], 8) == [0, 3]
```
