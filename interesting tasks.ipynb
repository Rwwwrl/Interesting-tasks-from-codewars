We all know about "balancing parentheses" (plus brackets, braces and chevrons) and even balancing characters that are identical.


***Kata***  
***Your challenge in this kata is to write a piece of code to validate that a supplied string is balanced.***

You must determine if all that is open is then closed, and nothing is closed which is not already open!

You will be given a string to validate, and a second string, where each pair of characters defines an opening and closing sequence that needs balancing.

You may assume that the second string always has an even number of characters.

*(Это нужно было решать с помощью стака)*

```python
from collections import deque


def is_balanced(source, caps):
    '''
    Проверить является ли строка сбалансированной,
    удалость решить O(n) 
    '''
    open_to_close = {}
    close_to_open = {}
    deq = deque()
    unique_deq = deque()
    for i in range(0, len(caps), 2):
        open_to_close[caps[i]] = caps[i + 1]
        close_to_open[caps[i + 1]] = caps[i]
    for char in source:
        if char in open_to_close:
            if open_to_close[char] == char:
                try:
                    last_value = unique_deq.pop()
                except IndexError:
                    last_value = ''
                    if last_value != char:
                        if last_value:
                            unique_deq.append(last_value)
                        unique_deq.append(char)
            else:
                deq.append(char)
            continue
            
        elif char in close_to_open:
            try:
                last_value = deq.pop()
            except IndexError:
                return False
            if close_to_open[char] != last_value:
                return False

    return len(deq) == 0 and len(unique_deq) == 0


assert is_balanced('-some value', '--') == False
assert is_balanced('-some value-', '--') == True
assert is_balanced('-+some value+-', '--++') == True
assert is_balanced('((Hello))', '()') == True
```
